```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
#import dataset
airbnb = pd.read_csv("result_Airbnb.csv")
zoopla = pd.read_csv("result_Zoopla.csv")
boundary = gpd.read_file(r'data/glasgowBoundaries/Community_Councils.shp')

#transfer to geodataframe
airbnb_g = gpd.GeoDataFrame(airbnb,geometry = gpd.points_from_xy(airbnb.lng,airbnb.lat),crs = 'EPSG:4326')
airbnb_g = airbnb_g.to_crs(27700)

zoopla_g = gpd.GeoDataFrame(zoopla,geometry = gpd.points_from_xy(zoopla.lon,zoopla.lat),crs = 'EPSG:4326')
zoopla_g = zoopla_g.to_crs(27700)
```

```python
airbnb_l = airbnb_g[airbnb_g['Dominant_Topic']=='Property Location']
airbnb_l = airbnb_l.reset_index(drop=True)
zoopla_l = zoopla_g[zoopla_g['Dominant_Topic']=='Property Location']
zoopla_l = zoopla_l.reset_index(drop=True)

airbnb_s = airbnb_g[airbnb_g['Dominant_Topic']=='Property Structure']
airbnb_s = airbnb_s.reset_index(drop=True)
zoopla_s = zoopla_g[zoopla_g['Dominant_Topic']=='Property Structure']
zoopla_s = zoopla_s.reset_index(drop=True)

airbnb_a = airbnb_g[airbnb_g['Dominant_Topic']=='Student Accommodation']
airbnb_a = airbnb_a.reset_index(drop=True)
zoopla_a = zoopla_g[zoopla_g['Dominant_Topic']=='Student Accommodation']
zoopla_a = zoopla_a.reset_index(drop=True)
```

```python
#zoopla
#plotting for check
plt.rcParams['axes.facecolor']='#cccccc'
ax = zoopla_l.plot(color = '#50ff7e', zorder=3, markersize = 1)
ax1 = zoopla_s.plot(ax = ax, color = '#aa50ff', zorder=2, markersize = 0.5)
ax2 = zoopla_a.plot(ax = ax1, color = 'blue', zorder=2, markersize = 0.5)
boundary.plot(ax = ax2, zorder = 1, color='w',edgecolor='k')
plt.xticks([])
plt.yticks([])
plt.grid(True, alpha=0.5)
# plt.title('Distribution of Zoopla Listings with Different Theme', fontsize=16)
plt.legend(['Location','Structure','Student Accommodation'])
plt.savefig('Zoopla Theme.png',dpi=300)

#airbnb
#plotting for check
plt.rcParams['axes.facecolor']='#cccccc'
ax = airbnb_l.plot(color = '#aa50ff', zorder=2, markersize = 0.5)
ax1 = airbnb_s.plot(ax = ax, color = '#50ff7e', zorder=2, markersize = 0.5)
ax2 = airbnb_a.plot(ax = ax1, color = 'blue', zorder=3, markersize = 0.5)
boundary.plot(ax = ax2, zorder = 1, color='w',edgecolor='k')
plt.xticks([])
plt.yticks([])
plt.grid(True, alpha=0.5)
# plt.title('Distribution of Airbnb Listings with Different Theme', fontsize=16)
plt.legend(['Location','Structure','Student Accommodation'])
plt.savefig('Airbnb Theme.png',dpi=300)
```
